```python
import pandas as pd
import numpy as np
```

```python
smartphones = pd.read_csv('flipkart_smartphones_data.csv')
smartphones_copy = smartphones.copy()
```

```python
smartphones.columns
```

```python
columns_to_drop = [
    'Mini USB Version', 'Hot SWAP Support', 'WAP', 'Games', 'International Warranty',
    'WAP Version', 'EDGE Features', 'GPRS Features', 'Important Apps', 'Keypad Type',
    'Bluetooth Version', 'Audio Jack', 'USB Type', 'SIM Size', 'Processor Core',
    'Warranty Summary', 'Domestic Warranty', 'Removable Battery', 'Touchscreen',
    'Primary Camera Available', 'Secondary Camera', 'Dual Camera Lens'
]
smartphones.drop(columns=[col for col in columns_to_drop if col in smartphones.columns], inplace=True)
```

```python
smartphones.info()
```

```python
smartphones.head(2)
```

```python
smartphones_df = smartphones.drop(columns=[col for col in smartphones.columns if smartphones[col].isna().sum() > 1000])
```

```python
smartphones_df.info()
```

```python
smartphones_df = smartphones_df[~smartphones_df['Product Name'].isna()]
smartphones_df = smartphones_df[smartphones_df['Product Name'] != 'Not Available']
```

```python
smartphones_df.info()
```

# Drop all Columns of no use

```python
# List of columns to drop based on relevance
columns_to_drop = [
    'Product URL',
    'In The Box',
    'Model Number',
    'Model Name',
    'Color',
    'Browse Type',
    'SIM Type',
    'Hybrid Sim Slot',
    'OTG Compatible',
    'Resolution Type',
    'Display Type',
    'Other Display Features',
    'Processor Brand',
    'Processor Type',
    'Primary Clock Speed',
    'Operating Frequency',
    'Primary Camera Features',
    'Secondary Camera Available',
    'Secondary Camera Features',
    'Flash',
    'Video Recording Resolution',
    'Digital Zoom',
    'Frame Rate',
    'Network Type',
    'Supported Networks',
    'Internet Connectivity',
    '3G',
    'Bluetooth Support',
    'Wi-Fi',
    'Wi-Fi Version',
    'Wi-Fi Hotspot',
    'NFC',
    'USB Connectivity',
    'GPS Support',
    'Smartphone',
    'User Interface',
    'SMS',
    'Sensors',
    'Other Features',
    'GPS Type',
    'Video Formats',
    'Battery Type',
    'Width',
    'Height',
    'Depth',
    'Quick Charging',
    'GPU',
    'Secondary Clock Speed',
    'HD Recording',
    'Full HD Recording',
    'Video Recording',
    'Speaker Phone',
    'Micro USB Port',
    'Micro USB Version',
    'Touchscreen Type',
    'Graphics PPI',
    'FM Radio'
]

# Drop the columns
df = smartphones_df.drop(columns=columns_to_drop)
```

```python
df.head(2)
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df.info()
```

```python
df_copy = df.copy()
```

```python
df = df.drop(columns=['Weight', 'Is Flipkart Assured'])
```

```python
df['RAM'] = df['RAM'].str.replace(' GB', '').astype(float, errors='ignore')  # Remove 'GB' and convert to float
df['Battery Capacity'] = df['Battery Capacity'].str.replace(' mAh', '').astype(float, errors='ignore')  # Remove 'mAh' and convert to float
```

```python
df['RAM'] = df['RAM'].fillna(df['RAM'].median())
df['Battery Capacity'] = df['Battery Capacity'].fillna(df['Battery Capacity'].median())
```

```python
# Rename columns to include units in the headers
df.rename(columns={'RAM': 'RAM (GB)', 'Battery Capacity': 'Battery Capacity (mAh)'}, inplace=True)

# Check the updated column names
print(df.columns)
```

```python
df.head()
```

```python
df = df.drop(columns=['Memory Card Slot Type'])
```

Extracting brand name from product name

```python
df['Brand Name'] = df['Product Name'].str.split(' ').str[0]
```

```python
df.head(2)
```

```python
df['Ratings Count'] = df['Number of Ratings and Reviews'].str.extract(r'(\d+),?\d*\sRatings')[0]
df['Reviews Count'] = df['Number of Ratings and Reviews'].str.extract(r'(\d+)\sReviews')[0]

# Fill NaN values with 0 and convert to integer
df['Ratings Count'] = df['Ratings Count'].fillna(0).astype(int)
df['Reviews Count'] = df['Reviews Count'].fillna(0).astype(int)
```

```python
df.head()
```

```python
df = df.drop(columns=['Number of Ratings and Reviews'])
```

```python
df.info()
```

```python
df['Internal Storage'] = df['Internal Storage'].str.replace(' GB', '').astype(int, errors='ignore')
```

```python
df = df.drop(columns=['Display Size (cm)'])
```

```python
df['Price Before Discount'] = df['Price Before Discount'].str.replace('₹', '').str.replace(',', '').astype(int)
```

```python
df.info()
```

```python
df['Internal Storage (GB)'] = df['Internal Storage'].rename('Internal Storage (GB)')
```

```python
df = df.drop(columns=['Internal Storage'])
```

```python
df.head(2)
```

```python
df = df[~df['Discount'].str.contains('Not Available', na=False)]

# Remove '% off' from 'Discount' column and convert to float
df['Discount'] = df['Discount'].str.replace('% off', '').astype(float)
```

```python
df.head(2)
```

```python
df = df.rename(columns={'Discount': 'Discount (%)'})
```

```python
df.to_csv('cleaned_smartphone_data.csv', index=False)
```
